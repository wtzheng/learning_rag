# 01 — 文档加载与探索

这个 notebook 展示文档加载后的结构、PDF 提取质量、metadata 示例。

```python
import sys
sys.path.insert(0, '..')
import logging
logging.basicConfig(level=logging.WARNING, format='%(levelname)s:%(message)s')

from src.loader import load_directory, load_document
from pathlib import Path
```

## 扫描目录

`load_directory()` 会遍历 `data/raw/` 并加载所有 PDF/TXT。

```python
docs = load_directory('../data/raw')
print(f'Loaded {len(docs)} document(s)')
for doc in docs:
    print(f'  - {doc.metadata["source"]} ({doc.metadata["total_chars"]} chars, type={doc.metadata["type"]})')
```

## 检查文档内容

看前 500 字符。

```python
for doc in docs:
    print(f'=== {doc.metadata["source"]} ===')
    print(doc.text[:500])
    print('...')
    print()
```

## Metadata 结构

```python
for doc in docs:
    print(f'{doc.metadata["source"]}: {doc.metadata}')
```
